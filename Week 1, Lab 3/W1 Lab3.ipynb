```python
import pandas as pd

# Replace the file path if needed
df = pd.read_csv('hourly_weather_10_days.csv')
df.head()
```

```python

print(df.info())
print(df.describe())
print(df.isna().sum())
```

```python
# 1. Temperature ki khali jagahon ko us ke average se bharo
df['temperature_C'] = df['temperature_C'].fillna(df['temperature_C'].mean())

# 2. Wind Speed ki khali jagahon ko us ke average se bharo
df['wind_speed_kmph'] = df['wind_speed_kmph'].fillna(df['wind_speed_kmph'].mean())

# 3. Pressure ki khali jagahon ko us ke average se bharo
df['pressure_hPa'] = df['pressure_hPa'].fillna(df['pressure_hPa'].mean())

# 4. Ab check karte hain ke kya koi khali jagah bachi hai?
# 5. Humidity ki khali jagahon ko us ke average se bharo
df['humidity_%'] = df['humidity_%'].fillna(df['humidity_%'].mean())

# 6. Visibility ki khali jagahon ko us ke average se bharo
df['visibility_km'] = df['visibility_km'].fillna(df['visibility_km'].mean())

# Ab check karte hain dobara
print(df.isna().sum())
```

```python
# 1. 'timestamp' column ko asli date aur waqt mein badlein
df['timestamp'] = pd.to_datetime(df['timestamp'])

# 2. Waqt mein se 'ghanta' (Hour) alag nikal kar naya column banayein
df['hour'] = df['timestamp'].dt.hour

# 3. Har ghante ke hisab se average humidity check karein
avg_humidity_by_hour = df.groupby('hour')['humidity_%'].mean()
print(avg_humidity_by_hour)
```

```python
# TODO: Extract temperature and wind_speed as NumPy arrays
import numpy as np

temp = df['temperature_C'].to_numpy()
wind = df['wind_speed_kmph'].to_numpy()
```

```python
# 1. Temperature list ko 10 din aur 24 ghanton ki matrix mein badlein
temp_matrix = temp.reshape((10, 24))

# 2. Har din ka Minimum temperature nikalein
daily_min = temp_matrix.min(axis=1)

# 3. Har din ka Maximum temperature nikalein
daily_max = temp_matrix.max(axis=1)

# 4. Har din ka Average (Mean) temperature nikalein
daily_mean = temp_matrix.mean(axis=1)

# Nateeja dekhne ke liye print karein
print("Har din ka Mean Temperature:\n", daily_mean)
```

```python
# 1. Poori matrix ka overall Mean (Average) nikalein
matrix_mean = temp_matrix.mean()

# 2. Poori matrix ka Standard Deviation (std) nikalein
matrix_std = temp_matrix.std()

# 3. Formula lagakar matrix ko normalize karein
normalized_temp = (temp_matrix - matrix_mean) / matrix_std

# Check karne ke liye pehli row print karke dekhein
print("Normalized Matrix ki Pehli Row:\n", normalized_temp[0])
```

```python
# 1. Shart (Condition) lagayein ke wind speed 15 se zyada ho
mask = wind > 15

# 2. Is mask ko original wind array par apply karke data filter karein
high_wind = wind[mask]

# 3. Nateeja check karne ke liye print karein
print("Tez hawaon (High Wind) ki kul ginti:", len(high_wind))
print("Tez hawaon ki chand readings:", high_wind[:5])
```

```python
# 1. Function ki definition (Machine tayaar karna)
def daily_summary(matrix):
    summary_list = []  # Har din ka result save karne ke liye khali list
    
    # Loop chalayein jo 0 se lekar 9 tak (10 dinon ke liye) chalega
    for day in range(10):
        # Us makhsoos din (row) ka min, max aur mean nikalein
        d_min = matrix[day].min()
        d_max = matrix[day].max()
        d_mean = matrix[day].mean()
        
        # Ek dictionary banayein jis mein is din ka data save ho
        day_dict = {
            'Day': day + 1,
            'Min_Temp': d_min,
            'Max_Temp': d_max,
            'Mean_Temp': d_mean
        }
        # Is dictionary ko apni bari list mein shamil karte jayein
        summary_list.append(day_dict)
        
    return summary_list  # Aakhir mein poori list wapas kar do

# 2. Apne function ko test karein (Machine ko chala kar dekhein)
summaries = daily_summary(temp_matrix)

# Pehle do dinon ki summary screen par dikhayein
print(summaries[:2])
```
