```python
# ============================================
# CELL 1: Imports, File Path & Read CSV
# ============================================
import pandas as pd
import numpy as np

file_path = r"C:\Users\ST\Downloads\handling_missing_data_replace-20260706T131744Z-3-001\handling_missing_data_replace\weather_data.csv"

df = pd.read_csv(file_path)
print("Original DataFrame:")
print(df)
```

```python
# ============================================
# CELL 2: Replacing single value with NaN
# ============================================
print("\n" + "="*50)
print("Replacing -99999 with NaN:")
print("="*50)
new_df = df.replace(-99999, value=np.nan)
print(new_df)
```

```python
# ============================================
# CELL 3: Replacing list with single value
# ============================================
print("\n" + "="*50)
print("Replacing [-99999, -88888] with 0:")
print("="*50)
new_df = df.replace(to_replace=[-99999, -88888], value=0)
print(new_df)
```

```python
# ============================================
# CELL 4: Replacing per column
# ============================================
print("\n" + "="*50)
print("Replacing per column with NaN:")
print("="*50)
new_df = df.replace({
    'temperature': -99999,
    'windspeed': -99999,
    'event': '0'
}, np.nan)
print(new_df)
```

```python
# ============================================
# CELL 5: Replacing using mapping
# ============================================
print("\n" + "="*50)
print("Replacing using mapping:")
print("="*50)
new_df = df.replace({
    -99999: np.nan,
    'no event': 'Sunny',
})
print(new_df)
```

```python
# ============================================
# CELL 6: Regex replacement
# ============================================
print("\n" + "="*50)
print("Regex replacement (remove letters):")
print("="*50)
new_df = df.replace({'temperature': '[A-Za-z]', 'windspeed': '[a-z]'}, '', regex=True)
print(new_df)
```

```python
# ============================================
# CELL 7: Create student scores DataFrame
# ============================================
print("\n" + "="*50)
print("Student Scores DataFrame:")
print("="*50)
df_students = pd.DataFrame({
    'score': ['exceptional', 'average', 'good', 'poor', 'average', 'exceptional'],
    'student': ['rob', 'maya', 'parthiv', 'tom', 'julian', 'erica']
})
print(df_students)
```

```python
# ============================================
# CELL 8: Replacing list with another list
# ============================================
print("\n" + "="*50)
print("Replacing score strings with numbers:")
print("="*50)
df_students_replaced = df_students.replace(['poor', 'average', 'good', 'exceptional'], [1, 2, 3, 4])
print(df_students_replaced)
```

```python
# ============================================
# CELL 9: Final message
# ============================================
print("\n" + "="*50)
print("All operations completed successfully!")
print("="*50)
```
