```python
import pandas as pd


pak_weather = pd.DataFrame({
    "city": ["Lahore", "Karachi"],
    "temperature": [32, 35],
    "humidity": [80, 75]
})


pak_weather_2 = pd.DataFrame({
    "city": ["Mianwali", "Multan"],
    "temperature": [40, 42],
    "humidity": [60, 55]
})

print("df1:\n", pak_weather)
print("\ndf2:\n", pak_weather_2)
```

```python
final_weather_clean = pd.concat([pak_weather, pak_weather_2], ignore_index=True)
final_weather_clean
```

```python
weather_with_labels = pd.concat([pak_weather, pak_weather_2], keys=["pak_jan", "pak_feb"])

weather_with_labels
```
