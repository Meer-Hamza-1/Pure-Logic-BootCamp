```python
import pandas as pd
import numpy as np

print("Libraries imported successfully!")
```

```python
# 1. Pehla table banaya (Roll No aur City)
data1 = {
    'roll_no': [1, 2, 3, 4],
    'city': ['Lahore', 'Karachi', 'Islamabad', 'Peshawar']
}
df1 = pd.DataFrame(data1)

# 2. Dosra table banaya (Roll No aur Marks)
# Note: Is mein hum ne jan booch kar roll_no 5 dala hai aur 4 ko chor diya hai taakay joins samajh sakein
data2 = {
    'roll_no': [1, 2, 3, 5],
    'marks': [85, 77, 78, 92]
}
df2 = pd.DataFrame(data2)

# Dono tables ko screen par dekhne ke liye
print("--- Table 1 (df1) ---")
print(df1)
print("\n--- Table 2 (df2) ---")
print(df2)
```

```python
# 'on=roll_no' ka matlab hai roll number ki bunyad par joro
# 'how=inner' ka matlab hai sirf common data uthao
inner_join_df = pd.merge(df1, df2, on='roll_no', how='inner')

# Result dekhne ke liye
inner_join_df
```

```python
# 'how=left' ka matlab hai pehle table (df1) ko pora rakho
left_join_df = pd.merge(df1, df2, on='roll_no', how='left')

# Result dekhne ke liye
left_join_df
```

```python
# 'how=right' ka matlab hai dosray table (df2) ko poora rakho
right_join_df = pd.merge(df1, df2, on='roll_no', how='right')

# Result dekhne ke liye
right_join_df
```

```python
# 'how=outer' ka matlab hai dono tables ka poora data ikotha kar do
outer_join_df = pd.merge(df1, df2, on='roll_no', how='outer')

# Result dekhne ke liye
outer_join_df
```

```python
# Hum Pandas ko bol rahay hain ke 'outer_join_df' mein se sirf wo row uthao jahan city 'Lahore' hai
lahore_df = outer_join_df[outer_join_df['city'] == 'Lahore']

# Result dekhne ke liye
lahore_df
```

```python
# Hum Pandas ko bol rahay hain ke 'outer_join_df' mein se sirf wo row uthao jahan city 'Lahore' hai
Pershawar_df = outer_join_df[outer_join_df['city'] == 'Peshawar']

# Result dekhne ke liye
Pershawar_df
```

```python
# Hum Pandas ko bol rahay hain ke 'outer_join_df' mein se sirf wo rows uthao jahan city ya to Lahore ho ya Peshawar
lahore_peshawar_df = outer_join_df[outer_join_df['city'].isin(['Lahore', 'Peshawar'])]

# Result dekhne ke liye
lahore_peshawar_df
```
