```python
import pandas as pd
import numpy as np

# File path
file_path = r"C:\Users\ST\Downloads\pandas_dataframe_basics-20260706T132210Z-3-001\pandas_dataframe_basics\weather_data.csv"
```

```python
df = pd.read_csv(file_path)
print("Original DataFrame:")
print(df)
```

```python
# Shape
print("Shape (rows, columns):", df.shape)

# First 3 rows
print("\nFirst 3 rows:")
print(df.head(3))

# Last 2 rows
print("\nLast 2 rows:")
print(df.tail(2))

# Slicing rows
print("\nRows 1 to 3:")
print(df[1:3])

# Column names
print("\nColumn names:")
print(df.columns)

# Access specific column
print("\nDay column:")
print(df['day'])

# Access multiple columns
print("\nDay and Temperature:")
print(df[['day', 'temperature']])
```

```python
# Max temperature
print("Max Temperature:", df['temperature'].max())

# Filter rows where temperature > 32
print("\nRows with temperature > 32:")
print(df[df['temperature'] > 32])

# Day with max temperature
print("\nDay with max temperature:")
print(df['day'][df['temperature'] == df['temperature'].max()])

# Full row with max temperature
print("\nFull row with max temperature:")
print(df[df['temperature'] == df['temperature'].max()])

# Standard deviation
print("\nStandard Deviation:", df['temperature'].std())

# Describe statistics
print("\nStatistical Summary:")
print(df.describe())
```

```python
# Pakistani weather data
pakistani_weather = pd.DataFrame({
    "city": ["karachi","lahore","peshawar"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
print("Pakistani Weather:")
print(pakistani_weather)

# US weather data
us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
print("\nUS Weather:")
print(us_weather)
```

```python
# Concatenate with ignore_index
df_concat = pd.concat([pakistani_weather, us_weather], ignore_index=True)
print("Concatenated (ignore_index):")
print(df_concat)

# Concatenate with keys
df_concat_keys = pd.concat([pakistani_weather, us_weather], keys=["pakistan", "us"])
print("\nConcatenated (with keys):")
print(df_concat_keys)

# Access US cities
print("\nUS cities data:")
print(df_concat_keys.loc["us"])

# Access Pakistan cities
print("\nPakistan cities data:")
print(df_concat_keys.loc["pakistan"])
```

```python
temperature_df = pd.DataFrame({
    "city": ["karachi","lahore","peshawar"],
    "temperature": [32,45,30],
}, index=[0,1,2])
print("Temperature DataFrame:")
print(temperature_df)

windspeed_df = pd.DataFrame({
    "city": ["lahore","karachi"],
    "windspeed": [7,12],
}, index=[1,0])
print("\nWindspeed DataFrame:")
print(windspeed_df)

# Concatenate temperature and windspeed
df_combined = pd.concat([temperature_df, windspeed_df], axis=1)
print("\nCombined Temperature and Windspeed:")
print(df_combined)
```

```python
s = pd.Series(["Humid","Dry","Rain"], name="event")
print("Series:")
print(s)

df_with_series = pd.concat([temperature_df, s], axis=1)
print("\nDataFrame with Series added:")
print(df_with_series)
```

```python
# Read weather_by_cities.csv
file_path_cities = r"C:\Users\ST\Downloads\pandas_dataframe_basics-20260706T132210Z-3-001\pandas_dataframe_basics\weather_by_cities.csv"

try:
    df_cities = pd.read_csv(file_path_cities)
    print("Weather by Cities DataFrame:")
    print(df_cities)

    # Group by city
    g = df_cities.groupby("city")
    print("\nGroupBy object:", g)

    # Iterate through groups
    print("\nIterating through groups:")
    for city, data in g:
        print(f"\ncity: {city}")
        print(f"data:\n{data}")

    # Get specific group
    print("\nGetting group 'paris':")
    print(g.get_group('paris'))

    # Min values per group
    print("\nMinimum values per group:")
    print(g.min())

    # Mean values per group
    print("\nMean values per group:")
    print(g.mean(numeric_only=True))

    # Size of each group
    print("\nSize of each group:")
    print(g.size())

except FileNotFoundError:
    print("weather_by_cities.csv not found. Skipping group by section.")
```

```python
# Basic merge
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
print("df1:")
print(df1)

df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
print("\ndf2:")
print(df2)

# Merge on city
df3 = pd.merge(df1, df2, on="city")
print("\nMerged df1 and df2:")
print(df3)
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})
print("df1:")
print(df1)

df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})
print("\ndf2:")
print(df2)

# Inner join
df_inner = pd.merge(df1, df2, on="city", how="inner")
print("\nInner Join:")
print(df_inner)

# Outer join
df_outer = pd.merge(df1, df2, on="city", how="outer")
print("\nOuter Join:")
print(df_outer)

# Left join
df_left = pd.merge(df1, df2, on="city", how="left")
print("\nLeft Join:")
print(df_left)

# Right join
df_right = pd.merge(df1, df2, on="city", how="right")
print("\nRight Join:")
print(df_right)
```

```python
# With indicator
df_indicator = pd.merge(df1, df2, on="city", how="outer", indicator=True)
print("Outer Join with Indicator:")
print(df_indicator)

# With suffixes
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
print("\ndf1 with humidity:")
print(df1)

df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
print("\ndf2 with temperature & humidity:")
print(df2)

df_suffix = pd.merge(df1, df2, on="city", how="outer", suffixes=('_first','_second'))
print("\nMerge with suffixes:")
print(df_suffix)
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1.set_index('city', inplace=True)
print("df1 with city as index:")
print(df1)

df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2.set_index('city', inplace=True)
print("\ndf2 with city as index:")
print(df2)

# Join dataframes
df_join = df1.join(df2)
print("\nJoined DataFrames:")
print(df_join)
```
