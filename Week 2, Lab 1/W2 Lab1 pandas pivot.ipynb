```python
import pandas as pd
import numpy as np
```

```python
# 'pd.read_csv' ka matlab hai CSV file ko parho aur use aik table (DataFrame) mein badlo
df = pd.read_csv("weather.csv")

# Sirf 'df' likhne se wo poora table screen par show ho jata hai
df
```

```python
# Hum Python ko bol rahay hain: City ko side par rakho aur Date ko upar columns bana do
pivoted_df = df.pivot(index='city', columns='date')

# Naye badlay huay table ko dekhne ke liye
pivoted_df
```

```python
# Is dafa hum ne aakhir mein 'values="humidity"' ka izafa kiya hai
humidity_df = df.pivot(index='city', columns='date', values='humidity')

# Naye filtered table ko dekhne ke liye
humidity_df
```

```python
# `/content/drive/...` mita kar sirf file ka naam likhein
df2 = pd.read_csv("weather2.csv")

# Data ko dekhne ke liye
df2
```

```python
# Teesri weather file ko load kiya jismein pure saal ka data hai
df3 = pd.read_csv("weather3.csv")

# Data ko dekhne ke liye
df3
```

```python
# pd.Grouper(freq='ME') ka matlab hai Month-End ke hisab se group karo
monthly_summary = df3.pivot_table(index=pd.Grouper(freq='ME', key='date'), columns='city')

# Final result dekhne ke liye
monthly_summary
```
